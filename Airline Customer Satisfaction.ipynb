```python
# Importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing the dataset
dataset = pd.read_csv("Invistico_Airline.csv")
dataset
```

```python
# Checking for missing variables 
dataset.isnull().sum()
```

```python
# Remove rows with missing values and create a clean dataset
data= dataset.dropna(subset=['Arrival Delay in Minutes'])

# Print shape before and after cleaning
print(f"Original dataset shape: {dataset.shape}")
print(f"Cleaned dataset shape: {data.shape}")

# Verify no missing values remain
print("\nMissing values after cleaning:")
print(data.isnull().sum()['Arrival Delay in Minutes'])
```

```python
# Checking the  descriptive statistics
data.describe()
```

```python
# Checking the datatype
data.dtypes
```

```python
# Calculate the overall customer satisfaction rate
satisfaction_rate = data['satisfaction'].value_counts(normalize=True) * 100
print(satisfaction_rate)

# Visualize the satisfaction rate
plt.figure(figsize=(8, 6))
sns.countplot(x='satisfaction', data=data)
plt.title('Overall Customer Satisfaction Rate')
plt.xlabel('Satisfaction')
plt.ylabel('Count')
plt.show()
```

```python
# Analyzing Satisfaction by Demographics

# Analyze satisfaction by gender
satisfaction_by_gender = data.groupby('Gender')['satisfaction'].value_counts(normalize=True) * 100
print(satisfaction_by_gender)

# Visualize satisfaction by gender
plt.figure(figsize=(10, 6))
sns.countplot(x='satisfaction', hue='Gender', data=data)
plt.title('Satisfaction by Gender')
plt.xlabel('Satisfaction')
plt.ylabel('Count')
plt.show()
```

```python
# Analyzing satisfaction by age group
data['Age Group'] = pd.cut(data['Age'], bins=[18, 24, 34, 44, 54, 64], labels=['18-24', '25-34', '35-44', '45-54', '55-64'])

satisfaction_by_age_group = data.groupby('Age Group')['satisfaction'].value_counts(normalize=True) * 100
print(satisfaction_by_age_group)

# Visualize satisfaction by age group
plt.figure(figsize=(10, 6))
sns.countplot(x='satisfaction', hue='Age Group', data=data)
plt.title('Satisfaction by Age Group')
plt.xlabel('Satisfaction')
plt.ylabel('Count')
plt.show()
```

```python
# Selecting only numeric columns for correlation calculation
numeric_data = data.select_dtypes(include=['int64', 'float64'])

# Calculate the correlation between columns
corr_matrix = numeric_data.corr()

# Visualize the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Identify the top factors correlated with customer satisfaction
# Ensure 'Satisfaction' column is numeric
satisfaction_column = pd.to_numeric(data['satisfaction'], errors='coerce')

# Calculate correlation between 'Satisfaction' and other numeric columns
corr_with_satisfaction = numeric_data.apply(lambda x: x.corr(satisfaction_column))


```

```python
# Identify top factors correlated with customer satisfaction
top_factors = corr_with_satisfaction.sort_values(ascending=False)[1:6]
print(top_factors)
```
